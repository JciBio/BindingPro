```python
import numpy as np
import scipy.io as sio
from sklearn.model_selection import KFold
import tensorflow as tf


# mnist = input_data.read_data_sets('MNIST_data',one_hot=True)
# print(mnist)



aa=[0,1]
bb=[1,0]
# #从MNIST数据集中筛选出5000条数据用作测试
# X_train,Y_train = mnist.train.next_batch(50000)
# #从MNIST数据集中筛选出200条数据用作测试
# X_test,Y_test = mnist.test.next_batch(1000)

def Knn(X_train,X_test,Y_train,Y_test):
    #图输入
    train2_X = tf.placeholder("float",[None,460])
    test2_X = tf.placeholder("float",[460])

    #使用L1距离计算KNN距离计算
    # tf.neg()给xte中的每个元素取反，这样可以使用tf.add()
    distance = tf.reduce_sum(tf.abs(tf.add(train2_X,tf.negative(test2_X))),reduction_indices=1)

    #预测：取得最近的邻居节点
    # 从distance tensor中寻找最小的距离索引
    pred = tf.argmin(distance,0)

    accuracy = 0
    TP = 0
    FP = 0
    TN = 0
    FN = 0
    #变量初始化
    init = tf.global_variables_initializer()

    #启动图
    with tf.Session() as sess:
        sess.run(init)
        #遍历测试数据集
        for i in range(len(X_test)):
            #获取最近的邻居节点
            nn_index = sess.run(pred,feed_dict={train2_X:X_train,test2_X:X_test[i,:]})
            
#             print("最近的邻居节点nn_indext:", nn_index)
            #print("邻居节点的值:", Y_train[nn_index])
#             print("该节点的真实类别:", np.argmax(Y_train[nn_index]))
#             print('')
            
            #print(np.argmax(Y_train[nn_index]),np.argmax(Y_test[i]))
            #获取最近的邻居节点的类别标签，并将其与该节点的真实类别标签进行比较
            #print("测试数据",i,"预测分类:",np.argmax(Y_train[nn_index]),"真实类别:",np.argmax(Y_test[i]))
            #计算准确率
            if np.argmax(Y_train[nn_index]) == np.argmax(Y_test[i]):
                accuracy += 1./len(X_test)
                
            if np.argmax(Y_train[nn_index]) == np.argmax(aa):
                
                if np.argmax(Y_train[nn_index]) ==np.argmax(Y_test[i]):
                    TP += 1
                else:
                    FP += 1
                    
            if np.argmax(Y_train[nn_index]) == np.argmax(bb):
                if np.argmax(Y_train[nn_index]) ==np.argmax(Y_test[i]):
                    TN += 1
                else:
                    FN += 1
            
        
        print( "TP:  ",TP)
        print( "FP:  ",FP)
        print( "TN:  ",TN)
        print( "FN:  ",FN)
        print("分类准确率为:",accuracy)
        print('')
        
        
#load benchmark dataset
data = sio.loadmat('PDNA-224-PSSM-11.mat')
X = data['data']
Y = data['target']
X = X.reshape(57348,-1)
#loo = LeaveOneOut()
kf = KFold(n_splits=5)
kf.get_n_splits(X)
for train_index, test_index in kf.split(X):
    X_train, X_test = X[train_index], X[test_index]
    Y_train, Y_test = Y[train_index], Y[test_index]
    print("len(X_train)={},len(Y_test)={}".format(len(X_train),len(Y_test)))
    print("YES!")
    Knn(X_train,X_test,Y_train,Y_test)
    
    
    
    
```
