```python
#将PDNA-62.fasta文件分割成单独的FASTA文件
for seq_record in SeqIO.parse("PDNA-62.fasta","fasta"):
    #print(seq_record.id)
    seq_ID=str(seq_record.id)
    outfilename=seq_ID.replace(":", "_")
    outfile=outfilename+'.fasta'
    #print(outfile)
    if len(seq_record.seq)>39:
        with open(outfile,"w") as f:       
            f.write(str(seq_record.id)+'\n' )
            f.write(str(seq_record.seq))
            f.close()
            print(outfile)
            #print(seq_record.seq)
            
        
```

```python
#将每个蛋白质序列进行滑窗处理，窗口大小为13
with open("1A02F_F_out.txt","w") as f:
     for seq_record in SeqIO.parse("1A02F_F.fasta","fasta"):
            f.write(str(seq_record.id)+'\n' )
            for i in range(len(seq_record.seq) - 13) :
                    f.write(str(seq_record.seq[i:i+13])+'\n' )   #first 13base positions
                    
                    
```
